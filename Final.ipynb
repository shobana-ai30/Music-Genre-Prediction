```python
#Genre Score - Baseline Model heuristics: A brute force approach
from sklearn import metrics
from sklearn.model_selection import train_test_split

#X = df.loc[:, ~df.columns.isin(['track_id', 'release', 'genre_id',	'genre_top',	'genre_label', 'artist_name',	'release'	,'genre_label_firstvalue'])]
X = df[['acousticness', 'danceability', 'instrumentalness', 'energy', 'valence', 'speechiness', 'liveness', 'tempo']]
X = X.to_numpy()
Y = df[['genre_label_firstvalue']]
Y = Y.to_numpy()

xtrain , xtest, ytrain, ytest = train_test_split(X, Y , test_size = 0.20)

p = []
score = {2:0, 3:0, 4:0, 5:0, 8:0, 10:0, 12:0, 15:0, 17:0, 21:0, 38:0, 1235:0}
#Classification Heuristics for HipHop
for index,row in df.iterrows():
    for k in score:
        score[k] = 0
    if (row["acousticness"] < 0.35):
        for key in score:
            score[key] = score[key] + 1
    if (row["acousticness"] > 0.35):
        for key in score:
            if key!=15:
                score[key] = score[key] + 1
    if (row["acousticness"] >0.42):
        for key in score:
            if key not in (15,21):
                score[key] = score[key] + 1
    if (row["acousticness"] >0.49):
        for key in score:
            if key not in (10,15,21):
                score[key] = score[key] + 1
    if (row["acousticness"] >=0.6):
        for key in score:
            if key not in (10,12,15,21,1235):
                score[key] = score[key] + 1
    if (row["acousticness"] >=0.61):
        for key in score:
            if key not in (10,12,15,21,38,1235):
                score[key] = score[key] + 1
    if (row["acousticness"] >=0.77):
        for key in score:
            if key not in (4,10,12,15,21,1235):
                score[key] = score[key] + 1
    if (row["acousticness"] >=0.82):
        for key in score:
            if key not in (3,4,10,12,15,17,21,1235):
                score[key] = score[key] + 1
    if (row["acousticness"] >=0.83):
        for key in score:
            if key not in (2,3,4,10,12,15,17,21,1235):
                score[key] = score[key] + 1
    if (row["acousticness"] >=0.98):
        for key in score:
            if key not in (2,3,4,8,10,12,15,17,21,1235):
                score[key] = score[key] + 1
    if (row["danceability"] < 0.32):
        for key in score:
            score[key] = score[key] + 1
    if (row["danceability"] > 0.32):
        for key in score:
            if key!=5:
                score[key] = score[key] + 1
    p.append(max(score,key=score.get))

#print("For all features")
print("Accuracy",metrics.accuracy_score(Y, p))
print("Confusion Matrix \n",metrics.confusion_matrix(Y, p))
```

```python
#Genre Score - Modified Baseline Heuristics
p = []
#Classification Heuristics for HipHop
for index,row in df.iterrows():
  if 0.5 < row["acousticness"] <= 0.6:
     p.append(10)
     continue
  if row["acousticness"] < 0.42:
     if row["acousticness"] < 0.4:
       p.append(15)
       continue
     p.append(21)
     continue
  if row["energy"] > 0.6:
    p.append(12)
    continue
  else:
    p.append(0)

#print("For all features")
print("Accuracy",metrics.accuracy_score(Y, p))
print("Confusion Matrix \n",metrics.confusion_matrix(Y, p))
```
